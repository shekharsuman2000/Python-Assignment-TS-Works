```python
import yfinance as yf
import sqlite3

# Read company names from file
with open("./data/companies.txt", "r") as f:
    companies = [line.strip() for line in f]

# Connect to database
conn = sqlite3.connect('finance.db')
c = conn.cursor()

# Create table if it doesn't exist
c.execute('''CREATE TABLE IF NOT EXISTS stock_prices
             (company, date, open, high, low, close, volume)''')

# Loop over each company
for company in companies:
    # Get historical data from Yahoo Finance API
    ticker = yf.Ticker(company)
    data = ticker.history(period="max")
    
    # Store data in database
    for index, row in data.iterrows():
        values = (company, str(index.date()), row['Open'], row['High'], row['Low'], row['Close'], row['Volume'])
        c.execute("INSERT INTO stock_prices VALUES (?, ?, ?, ?, ?, ?, ?)", values)
#
# Commit changes and close connection
conn.commit()
conn.close()
```

```python
import sqlite3

# Connect to database
conn = sqlite3.connect('finance.db')
c = conn.cursor()

# Check if the table exists
table_name = 'stock_prices'
c.execute(f"SELECT name FROM sqlite_master WHERE type='table' AND name='{table_name}'")
result = c.fetchone()

if result:
    print(f"Table {table_name} exists in the database")
else:
    print(f"Table {table_name} does not exist in the database")

# Close the connection
conn.close()
```
